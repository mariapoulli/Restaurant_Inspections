##### Reformat the data

```python
# Create subsets of the datasets consisting of only the columns of interest
insp_sub = insp[['INSPECTION_TYPE', 'JOB_PROGRESS', 'BBL', 'BOROUGH', 'INSPECTION_DATE', 'RESULT', 'APPROVED_DATE']]
apps_sub = apps[['Seating Interest (Sidewalk/Roadway/Both)', 'Restaurant Name', 'Sidewalk Dimensions (Area)',
                 'Roadway Dimensions (Area)', 'Approved for Sidewalk Seating', 'Approved for Roadway Seating', 
                 'Qualify Alcohol', 'Landmark District or Building', 'healthCompliance_terms', 'Time of Submission', 'BBL']]

# merge the datasets
data = pd.merge(insp_sub, apps_sub, on='BBL')

# define column names
column_names = ['Inspection_type', 'Job_progress', 'BBL', 'Borough', 'Inspection_date', 'Result', 'Approved_date',
                'Seating_interest', 'Restaurant_name', 'Sidewalk_dim_area', 'Roadway_dim_area', 'App_for_sw_seating', 
                'App_for_road_seating', 'Quality_alcohol', 'Land_dist_or_build', 'Health_comp_terms', 'Submission_date']

# define the column order
column_order= ['BBL', 'Restaurant_name', 'Borough', 'Land_dist_or_build', 'Health_comp_terms', 'Quality_alcohol',
               'Seating_interest', 'Sidewalk_dim_area', 'Roadway_dim_area', 'App_for_sw_seating', 'App_for_road_seating',
               'Inspection_type', 'Inspection_date', 'Job_progress', 'Result', 'Approved_date', 'Submission_date']

data.columns = column_names # rename the columns
data = data.reindex(columns=column_order) # reorder the columns
data.head(20) # explore the data
```

##### Check the column types

```python
data.dtypes
```

##### Fix the column types

```python
data['Inspection_date'] = pd.to_datetime(data['Inspection_date']).dt.date.astype('datetime64')
data['Approved_date'] = pd.to_datetime(data['Approved_date']).dt.date.astype('datetime64')
data['Submission_date'] = pd.to_datetime(data['Submission_date']).dt.date.astype('datetime64')
data['Sidewalk_dim_area'] = pd.to_numeric(data['Sidewalk_dim_area'], errors='coerce')
data['Roadway_dim_area'] = pd.to_numeric(data['Sidewalk_dim_area'], errors='coerce')
data.dtypes
```

##### Confirm that values are missing in the Sidewalk_dim_area and Roadway_dim_area because these restaurants didn't apply for the sidewalk and roadway seating.

```python
print('Roadway dim area missing:', data.isna().sum()['Roadway_dim_area'], 
      '\nNumber of restaurants that did not apply for road seating:', data['App_for_road_seating'].value_counts()['no'], 
      '\nSideway dim area missing:', data.isna().sum()['Sidewalk_dim_area'], 
      '\nNumber of restaurants that did not apply for sidewalk seating:', data['App_for_sw_seating'].value_counts()['no'])
```

##### Replace the missing values with zeros.

```python
data[['Sidewalk_dim_area', 'Roadway_dim_area']] = data[['Sidewalk_dim_area', 'Roadway_dim_area']].replace(np.NaN, 0)
data.isna().sum()
```

###### Nearly 80% of the initial rodent inspections are in Manhattan, the borough with the biggest number of restaurants.

##### Most of the restaurants in Manhattan pass the inspection, often the inital one, leading to these restaurants being inspected only once.<br> Restaurants in Queens proportionately fail for other reasons much more frequently than the restaurants in the other boroughs.<br> Brooklyn restaurants require more monitoring relative to the restaurants in other boroughs.

```python
data.groupby('Result')['Borough'].value_counts(normalize=True).unstack().plot(kind='bar')
```

##### Inspection type trends over time

```python
a = pd.DataFrame(data.groupby('Inspection_date')['Inspection_type'].value_counts().unstack())
a = a[a.index>'2008-01-01']

fig, ax = plt.subplots(5,1)
fig.set_size_inches(15, 10)
fig.subplots_adjust(hspace=.5)

ax[0].plot(a.index,a['BAIT'], label='BAIT')
ax[0].set_title('BAIT')

ax[1].plot(a.index,a['CLEAN_UPS'], label='CLEAN_UPS')
ax[1].set_title('CLEAN_UPS')

ax[2].plot(a.index,a['Compliance'], label='Compliance')
ax[2].set_title('Compliance')

ax[3].plot(a.index,a['Initial'], label='Initial')
ax[3].set_title('Initial')

ax[4].plot(a.index,a['STOPPAGE'], label='STOPPAGE')
ax[4].set_title('STOPPAGE')
```
